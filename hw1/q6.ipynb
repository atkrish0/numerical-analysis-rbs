```python
# using the quasi monte carlo submodule from scipy

from scipy.stats import qmc

sobol_gen = qmc.Sobol(d=2)
sobol_points = sobol_gen.random_base2(m=8)

sobol_points = sobol_points[:150]

plt.scatter(sobol_points[:,0], sobol_points[:,1], color='violet', label='sobol points')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
# first integral

n = 1000000

x = np.random.uniform(0, 1, n)
function_values = np.exp(np.exp(x))
integral_estimate = np.mean(function_values)

print(f"Estimated value of the integral: {integral_estimate}")
```

```python
# second integral

n = 1000000
L = 100

x = np.random.uniform(0, L, n)
function_values = x * (1 + x**2)**(-2)
integral_estimate = np.mean(function_values) * L

print(f"Estimated value of the integral: {integral_estimate}")
```

```python
# third integral

n = 1000
m = 1000

a, b = -2, 2
c, d = 0, 1

x_samples = np.random.uniform(a, b, n)
y_samples = np.random.uniform(c, d, m)

function_sum = 0

for i in range(n):
    for j in range(m):
        function_sum += np.exp((x_samples[i] + y_samples[j])**2)

integral_estimate = function_sum
print(f"Estimated value of the integral: {integral_estimate}")
```

```python
# fourth integral

n = 1000
m = 1000
L = 10

x_samples = np.random.uniform(0, L, n)
y_samples = np.random.uniform(0, L, m)
function_sum = 0

for i in range(n):
    for j in range(m):
        if y_samples[j] <= x_samples[i]:
            function_sum += np.exp(-(x_samples[i] + y_samples[j]))

integral_estimate = function_sum
print(f"Estimated value of the integral: {integral_estimate}")
```
