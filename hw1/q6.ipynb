```python
# first integral
n = 1000000

x = np.random.uniform(0, 1, n)
function_values = np.exp(np.exp(x))

I = np.mean(function_values)
print(I)
```

```python
# second integral
n = 1000000
L = 1000

x = np.random.uniform(0, L, n)
f = x * (1 + x**2)**(-2)

I = np.mean(f) * L
print(I)
```

```python
# third integral
n = 10000
s = 0

for i in range(n):
    x = np.random.uniform(-2, 2)
    y = np.random.uniform(0, 1)
    t = (x + 2) / 4
    f = 4 * np.exp((4 * t - 2 + y) ** 2) 
    s += f

I = s / n
print(I)
```

```python
n = 10000
Z = np.zeros(n)

for i in range(n):
    x = np.random.rand()
    y = np.random.rand() * x
    t = 1 / (1 + x)
    s = y / x
    f = np.exp(-(1/t - 1) * (1 + s)) * (1 / t**2) * abs(1 / t-1) 
    Z[i] = f

I = np.mean(Z)
print(I)
```
